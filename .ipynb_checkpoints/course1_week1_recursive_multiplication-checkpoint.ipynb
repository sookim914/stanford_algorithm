```python
# recurive multiplication using Karatsuba method 
## assuming we can only do one digit multiplication 
def karatsuba (x,y): 
    n = len(x)
    print(n,x,y)
    if n == 1: 
        return int(x)*int(y)
    else: 
        a = x[0:int(n/2)]
        b = x[int(n/2):int(n)]
        c = y[0:int(n/2)]
        d = y[int(n/2):int(n)]
        return (10**(n))*karatsuba(a,c) + (10**(n/2))*(karatsuba(a,d) + karatsuba(b,c)) + karatsuba(b,d)
```
